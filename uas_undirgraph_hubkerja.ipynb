##Menampilkan Simpul pada Graph

```
class graph :
  def __init__(self,gdict=None):
    if gdict is None:
      gdict = []
    self.gdict = gdict

#get the key of the dictionary
  def getVertices(self):
    return list(self.gdict.keys())

#create dictionary with graph elements
graph_elements = { "Direktur Utama" : ["Managing Director"],
                    "Managing Director" : ["Direktur Utama","Direktur Pemasasran","Direktur Operasional","Direktur Teknik","Direktur Keuangan","Direktur Personalia"],
                    "Direktur Pemasaran" : ["Managing Director","Manajemen Pemasaran"],
                    "Direktur Operasional" : ["Managing Director","Manajer Cost Control","Manajer Konstruksi"],
                    "Direktur Teknik" : ["Managing Director","Manajer Estimate","Manajer Logistik"],
                    "Direktur Keuangan" : ["Managing Director","Manajer Keuangan","Manajer Pembukuan"],
                    "Direktur Personalia" : ["Managing Director","Manajer Personalia dan Umum"]
                  }
              
g = graph(graph_elements)
print(g.getVertices())
```

##Menampilkan sudut pada Graph

```
class graph :
  def __init__(self,gdict=None):
    if gdict is None:
      gdict = {}
    self.gdict = gdict

  def edges(self):
    return self.findedges()

#find the distincts list of edges
  def findedges (self):
    edgename = []
    for vrtx in self.gdict:
      for nxtvrtx in self.gdict[vrtx]:
        if {nxtvrtx, vrtx} not in edgename:
          edgename.append({vrtx,nxtvrtx})
    return edgename


#create dictionary with graph elements
graph_elements = { "Direktur Utama" : ["Managing Director"],
                    "Managing Director" : ["Direktur Utama","Direktur Pemasasran","Direktur Operasional","Direktur Teknik","Direktur Keuangan","Direktur Personalia"],
                    "Direktur Pemasaran" : ["Managing Director","Manajemen Pemasaran"],
                    "Direktur Operasional" : ["Managing Director","Manajer Cost Control","Manajer Konstruksi"],
                    "Direktur Teknik" : ["Managing Director","Manajer Estimate","Manajer Logistik"],
                    "Direktur Keuangan" : ["Managing Director","Manajer Keuangan","Manajer Pembukuan"],
                    "Direktur Personalia" : ["Managing Director","Manajer Personalia dan Umum"]
                  }

g = graph(graph_elements)
print(g.edges())  
```

##Menggambarkan Undirected graph dgn networkx

```
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
vertices = range(1,15)
edges =  [(1,2),(2,3),(3,8),(3,9),(4,2),(4,10),(4,11),(5,2),(5,12),(5,13),(6,2),(6,14),(7,2),(7,15)]
G = nx.Graph()

G.add_nodes_from(vertices)
G.add_edges_from(edges)
pos=nx.spring_layout(G)

nx.draw_networkx_nodes(G,pos,nodelist=[1],label=True,node_color='g',node_size=1000)
nx.draw_networkx_nodes(G,pos,nodelist=[2],label=True,node_color='r',node_size=1000)
nx.draw_networkx_nodes(G,pos,nodelist=[3,4,5,6,7],label=True,node_color='c',node_size=1000)
nx.draw_networkx_nodes(G,pos,nodelist=[8,9,10,11,12,13,14,15],label=True,node_color='y',node_size=1000)

nx.draw_networkx_edges(G,pos,edges,width=3,alpha=0.5,edge_color='b')
labels={}
labels[1]=r'Dir.Utama'
labels[2]=r'Managing Director'
labels[3]=r'Dir.Operasional'
labels[4]=r'Dir.Teknik'
labels[5]=r'Dir.Keuangan'
labels[6]=r'Dir.Pemasaran'
labels[7]=r'Dir.Personalia'
labels[8]=r'Mnj. Cost Control'
labels[9]=r'Mnj. Konstruksi'
labels[10]=r'Mnj. Logistik'
labels[11]=r'Mnj. Estimate'
labels[12]=r'Mnj. Keuangan'
labels[13]=r'Mnj. Pembukuan'
labels[14]=r'Mnj. Pemasaran'
labels[15]=r'Mnj. Umum & Personalia'
nx.draw_networkx_labels(G,pos,labels,font_size=16)

plt.show() # displa
```
